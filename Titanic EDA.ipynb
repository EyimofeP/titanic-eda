```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#Setting Style
sns.set_style("darkgrid")
```

```python
titanic = pd.read_csv("titanic.csv")
```

```python
titanic.head(10)
```

Check Information about the df

```python
titanic.info()
# 891 rows, 15 columns
```

Total Amount of people that survived

```python
len(titanic[titanic["survived"] == 1])
```

Total Amount of People that didn't survived

```python
len(titanic[titanic["survived"] == 0])
```

```python
sns.countplot(x="survived", palette="prism", data=titanic);
# 0 = Not Survived, 1 = Survived
```

```python
plt.figure(figsize=[9,9])
plt.title("Distibution of Ages with People who Survived Involved")
sns.histplot(x="age", data=titanic,hue="survived",palette="magma", kde=True);
```

```python
plt.figure(figsize=[12,8])
plt.title("Distibution of People that Survived with Fare Included")
sns.scatterplot(x="fare", y="age", hue="survived", palette="plasma", data=titanic);
```
